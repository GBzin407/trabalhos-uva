```python
"""
GRUPO:
Gabriel de Oliveira Irineu - 1230113289
Lucas Faria de Oliveira
Mateus Teixeirão Pontes Nader
Pedro Pereira Passos de Alencar Sucupira
"""

import threading
import queue
import time

# Memória compartilhada
# O Lock protege o contador para evitar condição de corrida
contador_processados = 0
lock = threading.Lock()

# Fila compartilhada entre as threads (máximo 5 itens)
fila = queue.Queue(maxsize=5)

# Função do Produtor
def produtor():
    for i in range(10):
        item = f"dado_{i}"
        fila.put(item)  # bloqueia se a fila estiver cheia
        print(f"[PRODUTOR] > Colocou: {item}; Fila: {fila.qsize()}/5")
        time.sleep(0.5)

# Função do Consumidor
def consumidor():
    global contador_processados  # acessa a variável compartilhada
    for i in range(10):
        item = fila.get()  # bloqueia se a fila estiver vazia

        # Lock garante que só uma thread altera o contador por vez
        with lock:
            contador_processados += 1
            total = contador_processados

        print(f"[CONSUMIDOR] < Pegou: {item}; Total processado: {total}")
        time.sleep(1)

# Criando e iniciando as threads
t1 = threading.Thread(target=produtor)
t2 = threading.Thread(target=consumidor)

t1.start()
t2.start()

# Aguarda as duas threads terminarem
t1.join()
t2.join()

print(f"\n Fim! Total de itens processados: {contador_processados}")
```
